### 2-Nettoyage des dataframes

#### 2.1-Car_info

```python
useless_cols= ['Unnamed: 0']
car_info = car_info.drop(useless_cols, axis=1)
car_info.head()
```

#### 2.2-Renting_info

```python
Renting_info_copy= renting_info.copy()

#Suppression colonne id
useless_cols= ['rental_id','car_id', 'previous_ended_rental_id' ]
Renting_info_copy = Renting_info_copy.drop(useless_cols, axis=1)
Renting_info_copy.head()
```

```python
def create_interval_range(interval):
    if interval < 0:
        return 'retour en avance'
    if interval == 0:
        return '0'
    elif interval <= 15:
        return f'1m-15m'
    elif interval <= 30:
        return f'16m-30m'
    elif interval <= 45:
        return f'31m-45m'
    elif interval <= 60:
        return f'46m-60m'
    elif interval <= 120:
        return f'1h-2h'
    elif interval <= 180:
        return f'2h-3h'
    elif interval <= 240:
        return f'3h-4h'
    elif interval <= 300:
        return f'4h-5h'
    elif interval <= 360:
        return f'5h-6h'
    elif interval <= 420:
        return f'6h-7h'
    elif interval <= 480:
        return f'7h-8h'
    else:
        return f'>9h'

# Appliquer la fonction à la colonne 'interval' pour créer 'plage_interval'
Renting_info_copy['plage_interval'] = Renting_info_copy['delay_at_checkout_in_minutes'].apply(create_interval_range)

Renting_info_copy.head()
```

```python
median_checkout_time = Renting_info_copy['delay_at_checkout_in_minutes'].median()
average_checkout_time = Renting_info_copy['delay_at_checkout_in_minutes'].mean()
median_time_between_rentals = Renting_info_copy['time_delta_with_previous_rental_in_minutes'].median()
average_time_between_rentals = Renting_info_copy['time_delta_with_previous_rental_in_minutes'].mean()

fig = go.Figure()

# Indicateur pour le temps médian de check-out
fig.add_trace(go.Indicator(
    mode="number",
    value=median_checkout_time,
    title="Temps médian de Check-out",
    domain={'row': 0, 'column': 0}
))

# Indicateur pour la moyenne de check-out
fig.add_trace(go.Indicator(
    mode="number",
    value=average_checkout_time,
    title="Moyenne de Check-out",
    domain={'row': 0, 'column': 1}
))
# Indicateur pour le temps médian de check-out
fig.add_trace(go.Indicator(
    mode="number",
    value=median_time_between_rentals,
    title="Temps médian entre deux locations",
    domain={'row': 1, 'column': 0}
))

# Indicateur pour la moyenne de check-out
fig.add_trace(go.Indicator(
    mode="number",
    value=average_time_between_rentals,
    title="Moyenne entre deux locations",
    domain={'row': 1, 'column': 1}
))

fig.update_layout(
    grid={'rows': 2, 'columns': 2},
    template={'data': {'indicator': [{'mode': "number+delta+gauge"}]}
})

fig.show()
```

```python
min_delay.loc[:, "impacted_by_delay"] = (min_delay["delay_at_checkout_in_minutes"]
    - min_delay["time_delta_with_previous_rental_in_minutes"]) > 0

min_delay
```

```python
fig = px.histogram(min_delay, x = 'impacted_by_delay', color= 'checkin_type', text_auto =True).update_xaxes(categoryorder= 'total descending')
fig.update_layout(title="nombre de véhicule impacté par le delais",
                  legend_title_text="Type de connexion")
fig.show()
```

```python
to_keep = min_delay["impacted_by_delay"] ==True
impacted_cars = min_delay.loc[to_keep,:]
impacted_cars
```

```python
percentage_impacted = (len(impacted_cars) / len(Renting_info_copy)) * 100
print(percentage_impacted)
```

```python
to_keep = (min_delay["impacted_by_delay"] == True) & (min_delay['plage_interval'] == '1m-15m')
result= len(min_delay.loc[to_keep,:])
print(result)
```

```python
to_keep = (min_delay["impacted_by_delay"] == True) & (min_delay['plage_interval'] == '16m-30m')
result= len(min_delay.loc[to_keep,:])
print(result)
```

```python
interval = min_delay['plage_interval'].unique().tolist()
result_list=[]
pourcent_list=[]

for i in interval:
    to_keep = (min_delay["impacted_by_delay"] == True) & (min_delay['plage_interval'] == i)
    result= len(min_delay.loc[to_keep,:])
    pourcent= ((len(min_delay.loc[to_keep,:])/len(Renting_info_copy))*100)
    result_list.append(result)
    pourcent_list.append(pourcent)

data_dict={ "plage_interval": interval, "nombre_de_vehicules_impactés": result_list, "pourcentage_de_véhicules_impactés": pourcent_list}

result_df=pd.DataFrame(data_dict)
result_df['pourcentage_de_véhicules_impactés'] = result_df['pourcentage_de_véhicules_impactés'].round(2)
result_df.sort_values(by = 'nombre_de_vehicules_impactés', inplace = False)
result_df
```
